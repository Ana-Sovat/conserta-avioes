```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
folder = '/content/drive/MyDrive/conserta-avioes'
```

```python
benchmark = pd.read_csv(folder+"/benchmark_inicial_corrigido.csv")
benchmark
```

```python
benchmark = benchmark.iloc[4:]
benchmark
```

```python
finetuning = pd.read_csv(folder+"resultados_finetuning.csv")
finetuning
```

```python
finetuning = finetuning.rename(columns={'Modelo Base':'Modelo'})
```

```python
finetuning
```

```python
comparacao = pd.merge(benchmark,finetuning,how='right',on='Modelo',suffixes=('_inicial','_final'))
comparacao
```

```python
comparacao = comparacao[['Modelo','Learning Rate','Recall@1_inicial','Recall@1_final']]
comparacao
```

```python
comparacao_melhor = comparacao.loc[comparacao['Learning Rate']<5e-04]
```

```python
comp_6 = comparacao.loc[comparacao['Learning Rate']==5e-6]
comp_5 = comparacao.loc[comparacao['Learning Rate']==5e-5]
comp_4 = comparacao.loc[comparacao['Learning Rate']==5e-4]
```

## Gr찼ficos com Matplotlib

```python
plt.figure(figsize=[12.4, 4.8])
plot1 = plt.subplot(1, 2, 1,ylim=(0.89,1),xmargin=0.3)
plt.grid(axis='y')
plt.title("Learning Rate = 5e-06")
for i in range(comp_6.shape[0]):
  plt.plot(comp_6.iloc[i,2:],label=comp_6.iloc[i,0],marker='o')
  plt.annotate(text=f'{comp_6.iloc[i,3]:.4f}',xy=('Recall@1_final',comp_6.iloc[i,3]),textcoords='offset pixels',xytext=(7,-1),va='center')
plt.legend()

plt.subplot(1,2,2,ylim=(0.89,1),xmargin=0.3)
plt.grid(axis='y')
plt.title("Learning Rate = 5e-05")
for i in range(comp_5.shape[0]):
  plt.plot(comp_5.iloc[i,2:],label=comp_5.iloc[i,0],marker='o')
  plt.annotate(text=f'{comp_5.iloc[i,3]:.4f}',xy=('Recall@1_final',comp_5.iloc[i,3]),textcoords='offset pixels',xytext=(7,-1),va='center')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.grid(axis='y')
ax.set_title("Learning Rate = 5e-04")
ax.set_xmargin(0.5)
for i in range(comp_4.shape[0]):
  ax.plot(comp_4.iloc[i,2:],label=comp_4.iloc[i,0],marker='o')
  ax.annotate(text=f'{comp_4.iloc[i,3]:.4f}',xy=('Recall@1_final',comp_4.iloc[i,3]),textcoords='offset pixels',xytext=(7,-1),va='center')
plt.legend()
plt.show()
```
