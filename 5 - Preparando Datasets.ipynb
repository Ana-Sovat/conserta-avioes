```python
from tqdm import tqdm
import pandas as pd
import numpy as np
```

```python
folder = '/content/drive/MyDrive/conserta-avioes'
```

```python
def monta_dataset(dataset):
  dataset['Frase'] = dataset['Frase'].apply(lambda x:x.strip().upper())
  indices = dataset['Grupo'].unique()
  grupos = [dataset.loc[dataset['Grupo']==i,'Frase'].to_list() for i in indices]
  positivas = []
  for grupo in grupos:
    for i in range(5):
      for j in range(i+1,5):
        par = (grupo[i],grupo[j],1)
        positivas.append(par)
  negativas = []
  for i in range(len(grupos)):
    for j in range(10):
      par = (grupos[i][j%5],grupos[(i+1+j)%len(grupos)][j%5],0)
      negativas.append(par)
  todas = positivas+negativas
  novo_dataset = pd.DataFrame(todas,columns=['Sentence1','Sentence2','Label'])
  #novo_dataset['ClassLabel'] = novo_dataset['Label'].apply(lambda x:'positivo' if x==1 else 'negativo')
  return novo_dataset
```

```python
dataset_treino = pd.read_csv(f'{folder}/dataset_treino.csv')
dataset_treino.head()
```

```python
novo_treino = monta_dataset(dataset_treino)
novo_treino
```

```python
novo_treino.to_csv(folder+"/dataset_treino_pronto.csv",index=False)
```

```python
dataset_validacao = pd.read_csv(f'{folder}/dataset_validacao.csv')
dataset_validacao.head()
```

```python
novo_validacao = monta_dataset(dataset_validacao)
novo_validacao
```

```python
novo_validacao.to_csv(folder+"/dataset_validacao_pronto.csv",index=False)
```
