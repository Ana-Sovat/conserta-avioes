```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from google.colab import drive
from sklearn.cluster import DBSCAN
```

```python
drive.mount('/content/drive')
folder = '/content/drive/MyDrive/conserta-avioes'
```

```python
dados = pd.read_csv(folder+"/problemas.csv")
dados
```

```python
modelo = SentenceTransformer("multi-qa-mpnet-base-cos-v1")
modelo
```

```python
embeddings = modelo.encode(dados['DESAB'].tolist(),show_progress_bar=True)
```

```python
embeddings.shape
```

```python
dbscan = DBSCAN(eps=0.4,min_samples=2)
dbscan
```

```python
clusters = dbscan.fit(embeddings)
```

```python
clusters.labels_
```

```python
dados['clusters'] = clusters.labels_
dados
```

```python
dados.clusters.unique()
```

```python
dados['embeddings'] = embeddings.tolist()
dados
```

```python
dados.loc[dados['clusters']==3].iloc[:30]
```

```python
for i in dados['clusters'].unique():
  print(i, dados.loc[dados['clusters']==i,'DESAB'].iloc[0],len(dados.loc[dados['clusters']==i]))
```

```python
sem_cluster = dados.loc[dados['clusters']==-1]
sem_cluster.shape
```

```python
com_cluster = dados.loc[dados['clusters']!=-1]
```

```python
dedup = com_cluster.groupby('clusters').agg('first')
dedup = dedup.reset_index()
dedup
```

```python
problemas_dedup = pd.concat([sem_cluster,dedup])
problemas_dedup
```

```python
problemas_dedup.to_csv(folder+"/problemas_dedup.csv",index=False)
```
